```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta
import yfinance as yf
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
```

```python
class EMAcross(Strategy):
    n1 = 50
    n2 = 100
    
    def init(self):
        close = self.data.Close
        self.ema1 = self.I(ta.trend.ema_indicator, pd.Series(close), self.n1)
        self.ema2 = self.I(ta.trend.ema_indicator, pd.Series(close), self.n2)
    
    def next(self):
        if crossover(self.ema1, self.ema2):
            self.buy()
        elif crossover(self.ema2, self.ema1):
            self.sell()
    
```

```python
df = yf.download(tickers = 'M&M.NS', start = '2018-01-01')
```

```python
bt = Backtest(df, EMAcross, cash = 10000, commission = 0.002, exclusive_orders = True)
```

```python
output = bt.run()
```
